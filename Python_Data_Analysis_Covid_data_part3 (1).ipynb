```python
#Get Data & Packages
import pandas as pd
import datetime
```

To get the Data , you need to go back to part 2 and run the code there and EXPORT the dataframe as a CSV file using 

**DataFrame.to_csv("FileName.csv", sep=";", index=False)**

```python
#Import the data
covid_data = pd.read_csv('Covid_data_2020_16_April.csv'
                         ,sep = ";"
                         )
covid_data = covid_data.assign(date = pd.to_datetime(covid_data.date, format='%Y-%m-%d'))
```

**Basic Data Inspection**

```python
#print rows
covid_data.head()
```

```python
#print Column Names
covid_data.columns
```

```python
# get unique values: Country
pd.set_option('display.max_rows',180)
(covid_data.filter(['country'])
.drop_duplicates()
)
```

```python
#Get unique values: Date 
(covid_data.filter(['date'])
.drop_duplicates()
)
```

```python
# check variables: lat and long
(covid_data.filter(['long','lat'])
.agg(['min','max'])
)
```

```python
# Examine variables: Confirmed, dead, Recovered
(covid_data.filter(['confirmed','dead','recovered'])
.describe())
```

```python
# Count Missing Values
(covid_data.isnull()
.sum())
```

```python
#spot check important countrie
pd.set_option('display.max_rows',300)
(covid_data
.filter(['country','date','confirmed','dead','recovered'])
.groupby(['country','date'])
.agg('sum')
.query('date==datetime.date(2020, 4, 16)')
.query('country in ["US","India","China","France","Italy"]')
)
```
