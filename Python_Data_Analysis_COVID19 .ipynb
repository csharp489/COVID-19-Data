```python
#Importing Packages

import pandas as pd
```

```python
#read csv data file

url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
covid_data=pd.read_csv(url)
```

```python
covid_data.columns
```

```python
covid_data.head()
```

```python
# Rename Column Name
covid_data_label=covid_data.rename(
    columns= {
        'Province/State':'subregion',
        'Country/Region':'country',
        'Lat':'lat',
        'Long':'long'
    })
```

```python
#Reshape the data

# melt the data into 'tidy' shape
covid_data_label=(covid_data_label.melt(id_vars=['country','subregion','lat','long'],
                                        var_name='date_raw',
                                        value_name='confirmed'
                                       ))
```

```python
# to check the first case in the country's subregion
print(covid_data_label.loc[covid_data_label['confirmed']>0, : ].head())
```

```python
#convert dates

#Inspect dates
covid_data_label.filter(['date_raw'])
```

```python
# Test date Conversion

(covid_data_label.assign(date=pd.to_datetime(covid_data_label.date_raw, format='%m/%d/%y'))
                .filter(['date', 'date_raw', 'confirmed'])
                .groupby(['date', 'date_raw'])
                .agg('sum')
                .sort_values('date')
)
```

```python
#backup
covid_data_Backup=covid_data_label.copy()
```

```python
covid_data_label=covid_data_label.assign(date=pd.to_datetime(covid_data_label.date_raw, format="%m/%d/%y"))
```

```python
#sort and rearrange data
covid_data_label = (covid_data_label
               .filter(['country', 'subregion', 'date', 'lat', 'long', 'confirmed'])
               .sort_values(['country','subregion','lat','long','date'])
               )
```

```python
#check the first confirmed case of any country 
print(covid_data_label.loc[(covid_data_label['country']=='US') & (covid_data_label['confirmed']>0) , :])
```
